```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
file_path = 'C:/Users/ishan/Downloads/UNSW_NB15.csv'
data = pd.read_csv(file_path)

# Step 1: Handle missing values (if any)
data_cleaned = data.dropna()

# Step 2: Drop unnecessary columns (e.g., 'id')
data_cleaned = data_cleaned.drop(columns=['id'])

# Step 3: Feature Engineering - Byte Ratio and Packet Rate
data_cleaned['byte_ratio'] = data_cleaned['sbytes'] / (data_cleaned['sbytes'] + data_cleaned['dbytes'] + 1)
data_cleaned['packet_rate_src'] = data_cleaned['spkts'] / (data_cleaned['dur'] + 1)
data_cleaned['packet_rate_dst'] = data_cleaned['dpkts'] / (data_cleaned['dur'] + 1)

# Step 4: Encode categorical features ('proto', 'service', 'state', 'attack_cat')
label_encoder = LabelEncoder()
data_cleaned['proto_encoded'] = label_encoder.fit_transform(data_cleaned['proto'])
data_cleaned['service_encoded'] = label_encoder.fit_transform(data_cleaned['service'])
data_cleaned['state_encoded'] = label_encoder.fit_transform(data_cleaned['state'])
data_cleaned['attack_cat_encoded'] = label_encoder.fit_transform(data_cleaned['attack_cat'])

# Step 5: Drop original categorical columns after encoding
data_cleaned = data_cleaned.drop(columns=['proto', 'service', 'state', 'attack_cat'])

# Step 6: Scaling numerical columns
scaler = StandardScaler()
numerical_columns = ['dur', 'spkts', 'dpkts', 'sbytes', 'dbytes', 'rate', 'sttl', 'dttl', 'sload', 'dload', 
                     'sloss', 'dloss', 'sinpkt', 'dinpkt', 'sjit', 'djit', 'swin', 'dwin', 'tcprtt', 'synack', 
                     'ackdat', 'byte_ratio', 'packet_rate_src', 'packet_rate_dst']
data_cleaned[numerical_columns] = scaler.fit_transform(data_cleaned[numerical_columns])

# Step 7: Split the data into features (X) and label (y)
X = data_cleaned.drop(columns=['label'])
y = data_cleaned['label']

# Display the cleaned and preprocessed data
print(X.head())
print(y.head())
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
majority_class = data_cleaned[data_cleaned['label'] == 1]
minority_class = data_cleaned[data_cleaned['label'] == 0]

# Undersample the majority class to match the minority class
majority_class_undersampled = resample(majority_class, 
                                       replace=False,    # Sample without replacement
                                       n_samples=len(minority_class),  # Match minority class count
                                       random_state=42)  # Set a seed for reproducibility

# Combine minority class with the undersampled majority class
undersampled_data = pd.concat([minority_class, majority_class_undersampled])

# Shuffle the data to mix up the rows
undersampled_data = undersampled_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Separate the features and labels again
X_undersampled = undersampled_data.drop(columns=['label'])
y_undersampled = undersampled_data['label']

# Display the new label distribution
print(y_undersampled.value_counts())
```

```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import logging
import cProfile
import pstats

# Set up logging (only important info)
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')

# Check if GPU is available and set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Example data (replace with actual preprocessed and balanced data)
X_undersampled = np.random.rand(1000, 47)  # Use your actual features here
y_undersampled = np.random.randint(0, 2, size=1000)  # Use your actual labels here

# Feature names (from your dataset after preprocessing and encoding)
feature_names = ['dur', 'spkts', 'dpkts', 'sbytes', 'dbytes', 'rate', 'sttl', 'dttl', 'sload', 'dload', 
                 'sloss', 'dloss', 'sinpkt', 'dinpkt', 'sjit', 'djit', 'swin', 'stcpb', 'dtcpb', 'dwin', 
                 'tcprtt', 'synack', 'ackdat', 'smean', 'dmean', 'trans_depth', 'response_body_len', 
                 'ct_srv_src', 'ct_state_ttl', 'ct_dst_ltm', 'ct_src_dport_ltm', 'ct_dst_sport_ltm', 
                 'ct_dst_src_ltm', 'is_ftp_login', 'ct_ftp_cmd', 'ct_flw_http_mthd', 'ct_src_ltm', 
                 'ct_srv_dst', 'is_sm_ips_ports', 'byte_ratio', 'packet_rate_src', 'packet_rate_dst', 
                 'proto_encoded', 'service_encoded', 'state_encoded', 'attack_cat_encoded']

# Convert data to PyTorch tensors and move to GPU
X_undersampled = torch.tensor(X_undersampled).float().to(device)
y_undersampled = torch.tensor(y_undersampled).long().to(device)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_undersampled.cpu().numpy(), y_undersampled.cpu().numpy(), test_size=0.3, random_state=42)

# Convert the training and testing data back to PyTorch tensors and move to GPU
X_train = torch.tensor(X_train).float().to(device)
X_test = torch.tensor(X_test).float().to(device)
y_train = torch.tensor(y_train).long().to(device)
y_test = torch.tensor(y_test).long().to(device)

# Parameters for ACO
alpha = 0.4  # Influence of pheromone
beta = 0.6  # Influence of heuristic value
rho = 0.3  # Pheromone evaporation rate
n_ants = 50  # Number of ants
n_iterations = 150  # Max iterations
pheromone = torch.ones(X_train.shape[1]).to(device)  # Initialize pheromone on GPU

# Loss function
criterion = nn.CrossEntropyLoss()

# Profiling
profiler = cProfile.Profile()
profiler.enable()

# ACO process
logging.info('Starting ACO-based Feature Selection')
best_features = []
best_score = 0

for iteration in range(n_iterations):
    logging.info(f'Iteration {iteration + 1}/{n_iterations} started.')
    best_iteration_score = 0
    all_solutions = []
    all_scores = []

    for ant in range(n_ants):
        selected_features = []
        
        # Ants probabilistically select features based on pheromone and heuristic
        for feature in range(X_train.shape[1]):
            heuristic = torch.rand(1).item()  # Heuristic (can be based on some logic)
            probability = (pheromone[feature] ** alpha) * (heuristic ** beta)
            if torch.rand(1).item() < probability:
                selected_features.append(feature)
        
        if len(selected_features) > 0:
            selected_features = torch.tensor(selected_features).to(device)

            # Dynamically adjust the classifier based on the number of selected features
            classifier = nn.Linear(len(selected_features), 2).to(device)
            optimizer = torch.optim.Adam(classifier.parameters())

            # Subset the selected features for training
            X_train_subset = X_train[:, selected_features]
            X_test_subset = X_test[:, selected_features]

            # Training on selected features
            classifier.train()
            optimizer.zero_grad()
            outputs = classifier(X_train_subset)
            loss = criterion(outputs, y_train)
            loss.backward()
            optimizer.step()

            # Evaluation on test set
            classifier.eval()
            with torch.no_grad():
                predictions = classifier(X_test_subset)
                _, predicted = torch.max(predictions, 1)
                accuracy = accuracy_score(y_test.cpu(), predicted.cpu())
            
            all_solutions.append(selected_features)
            all_scores.append(accuracy)

    # Update pheromone levels based on solution quality
    for idx, solution in enumerate(all_solutions):
        score = all_scores[idx]
        for feature in solution:
            pheromone[feature] = (1 - rho) * pheromone[feature] + score * rho

    # Track the best solution in the current iteration
    best_ant_idx = np.argmax(all_scores)
    if all_scores[best_ant_idx] > best_score:
        best_score = all_scores[best_ant_idx]
        best_features = all_solutions[best_ant_idx].cpu().tolist()

    logging.info(f'Best score in iteration {iteration + 1}: {best_score:.4f} with features {best_features}')

# Stop profiling
profiler.disable()

# Map the selected feature indices to their names
selected_feature_names = [feature_names[i] for i in best_features]
logging.info(f'Final selected feature names: {selected_feature_names}')
logging.info(f'Best accuracy achieved: {best_score:.4f}')

# Output profiling results
stats = pstats.Stats(profiler).sort_stats('cumtime')
stats.print_stats(10)  # Print top 10 time-consuming function calls
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import logging

# Logging setup
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')

# Sample Dataset (You can replace this with your actual dataset)
# X_undersampled = Your preprocessed data after undersampling
# y_undersampled = Labels corresponding to your dataset

# Assuming you have preprocessed and undersampled data as tensors
# Step 1: Create a Logistic Regression model to use with ACO
class LogisticRegression(nn.Module):
    def __init__(self, input_size):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return torch.sigmoid(self.linear(x))

# ACO parameters
alpha = 0.4  # Pheromone importance
beta = 0.6   # Heuristic information importance
rho = 0.2    # Pheromone evaporation rate
num_ants = 50
num_iterations = 100

# Step 2: ACO-Based Feature Selection
def aco_feature_selection(X_train, y_train):
    num_features = X_train.shape[1]
    pheromone = np.ones(num_features)
    best_score = 0
    best_features = []
    
    for iteration in range(num_iterations):
        logging.info(f"Iteration {iteration + 1}/{num_iterations} started.")
        # Step 2.1: Each ant selects a subset of features
        ant_solutions = []
        for _ in range(num_ants):
            probabilities = (pheromone ** alpha) * ((1 / (np.arange(1, num_features + 1))) ** beta)
            probabilities /= probabilities.sum()
            selected_features = np.random.choice(np.arange(num_features), size=num_features//2, replace=False, p=probabilities)
            ant_solutions.append(selected_features)

        # Step 2.2: Evaluate the performance of the subset
        best_ant_score = 0
        best_ant_features = []
        for features in ant_solutions:
            X_train_subset = X_train[:, features]
            # Train the logistic regression model
            model = LogisticRegression(X_train_subset.shape[1])
            criterion = nn.BCELoss()
            optimizer = optim.Adam(model.parameters(), lr=0.01)

            # Convert to torch tensors
            X_train_subset_tensor = torch.tensor(X_train_subset, dtype=torch.float32)
            y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
            
            # Train the model
            for epoch in range(100):  # Training loop
                optimizer.zero_grad()
                outputs = model(X_train_subset_tensor)
                loss = criterion(outputs, y_train_tensor)
                loss.backward()
                optimizer.step()

            # Calculate the accuracy
            with torch.no_grad():
                predicted = (outputs > 0.5).float()
                accuracy = (predicted.eq(y_train_tensor)).sum() / y_train_tensor.size(0)
                if accuracy > best_ant_score:
                    best_ant_score = accuracy.item()
                    best_ant_features = features

        # Update pheromone values
        pheromone *= (1 - rho)
        pheromone[best_ant_features] += best_ant_score

        if best_ant_score > best_score:
            best_score = best_ant_score
            best_features = best_ant_features

        logging.info(f"Best score in iteration {iteration + 1}: {best_ant_score:.4f} with features {best_ant_features}")

    return best_features, best_score

# Step 3: Apply Random Forest using the ACO-selected features
def apply_random_forest(X, y, selected_features):
    X = X[:, selected_features]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Train Random Forest
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = rf_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    logging.info(f"Random Forest Accuracy with ACO-selected features: {accuracy:.4f}")

# Step 4: Run ACO and Random Forest
def run_aco_random_forest(X_undersampled, y_undersampled):
    # Convert your dataset to numpy arrays (if not already)
    X_undersampled_np = X_undersampled.numpy() if isinstance(X_undersampled, torch.Tensor) else X_undersampled
    y_undersampled_np = y_undersampled.numpy() if isinstance(y_undersampled, torch.Tensor) else y_undersampled

    logging.info("Starting ACO-based Feature Selection")
    selected_features, best_score = aco_feature_selection(X_undersampled_np, y_undersampled_np)
    
    logging.info(f"ACO completed with best score: {best_score:.4f} and selected features: {selected_features}")
    
    logging.info("Applying Random Forest with selected features")
    apply_random_forest(X_undersampled_np, y_undersampled_np, selected_features)

# Run the combined process
run_aco_random_forest(X_undersampled, y_undersampled)  # Assuming X_undersampled and y_undersampled are your dataset tensors
```
